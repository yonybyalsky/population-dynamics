$P_{n+1} = \gamma P_{n}$
<br>
$P_{1} = \gamma P_{0}$
<br>
$P_{2} = \gamma P_{1} = \gamma (\gamma P_{0}) = \gamma^{2} P_{0}$
<br>
$P_{3} = \gamma P_{2} = \gamma (\gamma^{2} P_{0}) = \gamma^{3} P_{0}$
<br>
......
<br>
$P_{n} = \gamma * P_{n-1} = \gamma * (\gamma^{n-1} * P_{0}) = \gamma^{n} * P_{0}$
<br>
הוכחה לפי שיטת אינדוקציה מתמטית:
<br>
$n=k$:
<br>
$P_{k} = \gamma^{k} * P_{0}$
<br>
$P_{k+1} = \gamma^{k + 1} * P_{0}$
<br>
$P_{k+1} = \gamma * P_{k} = \gamma * (\gamma^{k} * P_{0}) = \gamma^{k+1} * P_0$

ככל שהאוכלסייה גודלת, המשאב קטן. המודל שלי מתבסס על כך שקטינת המשאב מהווה השפעה שלילית על גדילת האוכלסייה
<br>
$\beta$ - מקדם ההשפעה השלילית של קטינת המשאב על גדילת האוכלסייה כמו שגמא הוא מקדם ההשפעה החיובית (במקרה שיש יותר ילודה ממוות) של גדילת האוכלסייה
<br>
בטא תלוי בכמה גורמים, כמו שגמא תלוי באחוזי הלידה והתמותה, בטא תלוי בכמות המשאב, כמות הצריכה שלו וכו', ועל סמך זה אנו קובעים את הערך שלו. 
<br>
כיוון ההשפעה השלילית של קטינת המשאב תהיה יותר אינטנסיבית מאשר ההשפעה החיובית ה$P_{n}$ יהיה בחזקת 2 
<br>
$P_{n+1} = \gamma * P_{n} - \beta * P_{n} ^ {2}$
<br>
המודל מתאים בעיקר לאוכלסיות גדולות (שמתחילות לפחות באזור האלפים) אם נרצה מודל לאוכלסיות קטנות כמו להקת ארנבים המונה בהתחלה 5 ארנבים המודל הראשון יהיה יותר מתאים, וניתן להשתמש בו עד שנגמר המשאב.

```python
#exercise 2
def population_over_time(initial_population, gamma, beta):
    population = initial_population
    while True:
        yield population
        population = population * gamma - (beta * (population ** 2))
```

```python
#exercise 2 - example
max_year = 100
year = 0
for rabbits in population_over_time(1000, 1.2, 0.000025):
    print(f'{year=}, {rabbits=}')
    year += 1
    
    if year == max_year:
        break
```
