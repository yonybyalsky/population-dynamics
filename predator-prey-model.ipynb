```python
import matplotlib.pyplot as plt
```

```python
def predator_prey_model(initial_r_p, initial_w_p, alpha, beta, gamma, delta):
    r_n = initial_r_p
    r_n_this_year = initial_r_p
    w_n = initial_w_p
    while True:
        yield r_n, w_n
        r_n = alpha * r_n - gamma * w_n * r_n
        w_n = beta * w_n * r_n_this_year - delta * w_n
        r_n_this_year = alpha * r_n_this_year - gamma * w_n * r_n_this_year
```

```python
max_year = 90
year = 0
for rabbits, wolfs in predator_prey_model(500, 500, 2, 1/350, 1/500, 0.1):
    print(rabbits, wolfs)
    year += 1
    
    if year == max_year:
        break
```

```python
def plot_predator_prey_model(func, max_year):
    year = 0
    rabbits_list = []
    wolf_list = []
    year_list = []
    for rabbits, wolfs in func:
        rabbits_list.append(rabbits)
        wolf_list.append(wolfs)
        year_list.append(year)
        year += 1    
        
        if year == max_year:
            break
    plt.plot(year_list, rabbits_list, label='rabbits_population')
    plt.plot(year_list, wolf_list, label='wolfs_population')
    plt.legend(loc='best')
```

```python
plot_predator_prey_model(predator_prey_model(500, 500, 2, 1/350, 1/500, 0.1), 50)
```
