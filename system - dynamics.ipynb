```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def population(pop_percent, gamma, max_year):
    year = 0
    while year < max_year:
        pop_percent *= gamma*(1-pop_percent)
        year += 1
    return pop_percent
```

```python
x = []
y = []
z = []
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for gamma in np.arange(0, 4, 0.4):
    for pop_percent in np.arange(0, 1, 0.1):
        for year in np.arange(0, 50, 1):
            y.append(population(pop_percent, gamma, year))
            x.append(gamma)
            z.append(year)
ax.scatter(x, y, z)
```

```python
#population system in year 75
%matplotlib notebook
x = []
y = []
for gamma in np.arange(0, 4, 0.04):
    for pop_percent in np.arange(0, 1, 0.01):
        y.append(population(pop_percent, gamma, 75))
        x.append(gamma)
plt.scatter(x, y)
```
